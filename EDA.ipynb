```python
# Replace the -1 values with nan value
df["Rating"] = df["Rating"].apply(lambda x: np.nan if x==-1 else x)

# Fill missing values with the mean of the distribution.
df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
```
